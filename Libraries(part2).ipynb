# Introduction

Matplotlib is a widely-used Python library for data visualization that provides a flexible and powerful way to create a variety of static, animated, and interactive plots. It is particularly useful for visualizing data in a clear and informative manner, making it easier to understand complex datasets and communicate insights effectively.

```python
import matplotlib.pyplot as plt
import numpy as np
```

# matplotlib.pyplot

```python
x =  np.arange(1, 6)
y = x**2
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x, y)
```

```python
plt.figure(figsize=(15, 3))
plt.plot(x, y)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y, 'r--')
axes2.plot(x, x**3, 'g--')
```

## Key Plot Basic Types

```python
# Line
plt.plot(x, y)
```

```python
# Bar
plt.bar(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bar Plot')

# Horizontal Bar, use barh
```

```python
# Scatter
plt.scatter(x, y)
```

```python
# Pie
labels = ['A', 'B', 'C', 'D', 'E']
sizes = [15, 30, 45, 10, 20]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
```

## Subplot

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')

plt.subplot(1, 2, 2)
plt.bar(x, y)
```

## Subplots

```python
fig, ax = plt.subplots(2,2)
```

```python
fig, ax = plt.subplots(2,2, figsize = [10,5])

ax[0, 0].plot(x, y, 'r--')
ax[0, 0].set_xlabel('X-axis')
ax[0, 0].set_ylabel('Y-axis')
ax[0, 0].set_title('Title')
ax[0, 1].bar(x, y)
ax[1, 0].scatter(x, y)
ax[1, 1].barh(x, y)

plt.tight_layout()
```

# Exercises

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Salaries.csv')
df.head(3)
```

```python
df.info()
```

```python
df['BasePay']
```

```python
df = df[df['BasePay'] != 'Not Provided']
```

```python
df_plots = df[['BasePay', 'Year', "JobTitle"]]
df_plots
```

```python
qtYear = df_plots.groupby('Year').size()  #count for each columns

plt.figure(figsize = (8,5))
plt.bar(qtYear.index, qtYear.values)
plt.xlabel('Year')
plt.ylabel('Quantity')
plt.title('Quantity by Year')
```

```python
salaryAvg = df.groupby('Year')['BasePay'].mean()

plt.figure(figsize = (8,5))
plt.plot(salaryAvg.index, salaryAvg.values)
plt.xlabel('Salary')
plt.ylabel('Year')
plt.title('Average by Year')
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10,5))

#bar
ax[0].bar(qtYear.index, qtYear.values)
ax[0].set_xlabel('Year')
ax[0].set_ylabel('Quantity')
ax[0].set_title('Quantity by Year')

#line
ax[1].plot(salaryAvg.index, salaryAvg.values)
ax[1].set_xlabel('Salary')
ax[1].set_ylabel('Year')
ax[1].set_title('Average by Year')

plt.tight_layout()
```
